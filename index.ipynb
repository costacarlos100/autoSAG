```python
#PROJETO DE AUTOMAÇÃO DO SISTEMA COMPRASNET.GOV
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver import Firefox
from selenium import webdriver
from functools import partial
from openpyxl import Workbook
from time import sleep
import locale

locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
print("+==============================================+")
print("+==         AUTOMATIZADOR SAG 1.0            ==+")
print("+==============================================+")
print("Preencha as informações abaixo: ")
iniVigencia = input("Inicio da Vigência: xx/xx/xxxx > ")
fimVigencia = input("Fim da Vigência:    xx/xx/xxxx > ")
descricaoDe = input("Descrição simples do Item: Ex:. microcomputador > ")
exercito    = input("Deseja filtrar somente por OM's: S/n > ")
estado      = input("Fornecedor de qual estado: Ex:. RJ, SP, MG > ")
valMedio    = input("Qual o valor médio do Produto: Ex:. 1000")

```

```python
driver = Firefox()
driver.get("http://sag.eb.mil.br")

wdw = WebDriverWait(driver, 100, poll_frequency=0.2)

def esperabutton(xpath, webdriver):
    objeto = webdriver.find_elements(By.XPATH, xpath)
    return bool(objeto)

#Fecha popup
sleep(7)
wdw.until(partial(esperabutton, '/html/body/div[2]/div/div/div[1]/button'))
driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[1]/button').click()
```

```python
#PARTE 1
sleep(5)
def retornaObjeto(esc, cam):
    wdw.until(partial(esperabutton, cam))
    if(esc == "text"):
        objeto = driver.find_element(By.XPATH, cam).text
    elif(esc == "click"):
        objeto = driver.find_element(By.XPATH, cam).click()
    elif(esc == "id"):
        objeto = driver.find_element(By.ID, cam).text

    return objeto

def mudaPontoFlutuante(xxpath):
    temporaria = xxpath.replace(",",".")
    temporaria = locale.atof(xxpath)
    return temporaria

caminho = "/html/body/div[1]/div/div/div[10]/div/div[2]/div/table/tbody/tr[2]/td[14]"
xxpath = mudaPontoFlutuante(retornaObjeto("text", caminho))


    
```

```python
from selenium.webdriver.common.action_chains import ActionChains
texto = driver.find_element(By.ID, 'example_info').text
qntItem = locale.atof(texto[19:21])
comeco = 1

while(comeco <= qntItem):
    caminho = f"/html/body/div[1]/div/div/div[10]/div/div[2]/div/table/tbody/tr[{comeco}]/td[14]"
    valorReal = mudaPontoFlutuante(retornaObjeto("text", caminho))
    if(valorReal >= valMedio):
        print(valorReal)

        cam = f"/html/body/div[1]/div/div/div[10]/div/div[2]/div/table/tbody/tr[{comeco}]/td[2]/a[1]"
        element = driver.find_element(By.XPATH, cam)
        driver.execute_script("arguments[0].scrollIntoView();", element)
        element.click()

        sleep(3)
        
        wdw.until(partial(esperabutton, caminho))
        handles = driver.window_handles 
        for h in handles: 
            driver.switch_to.window(h) 
        clicaButton = "//*[@id='ui-id-1']"
        retornaObjeto("click", clicaButton)

        qntOM = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[3]/form/div/div[1]/fieldset[2]/table/tbody/tr[2]/td[1]/input")
        qntOM = int(qntOM.get_attribute("value"))

        if(qntOM >= 1):

            nomeOM = "/html/body/div[1]/div/div[3]/form/table[1]/tbody/tr[2]/td[2]/input[2]"
            nomeOM = driver.find_element(By.XPATH, nomeOM)
            nomeOM = nomeOM.get_attribute("value")

            nLicita = "/html/body/div[1]/div/div[3]/form/table[2]/tbody/tr[2]/td[3]/input"
            nLicita = driver.find_element(By.XPATH, nLicita)
            nLicita = nLicita.get_attribute("value")

            descricao = "/html/body/div[1]/div/div[3]/form/table[4]/tbody/tr[2]/td/textarea"
            descricao = retornaObjeto("text", descricao)

            iniVig = "/html/body/div[1]/div/div[3]/form/div/div[1]/table[1]/tbody/tr[2]/td[3]/input[1]"
            iniVig = driver.find_element(By.XPATH, iniVig)
            iniVig = iniVig.get_attribute("value")

            fimVig = "/html/body/div[1]/div/div[3]/form/div/div[1]/table[1]/tbody/tr[2]/td[3]/input[2]"
            fimVig = driver.find_element(By.XPATH, fimVig)
            fimVig = fimVig.get_attribute("value")

            clicarFornecedor = '//*[@id="ui-id-4"]'
            retornaObjeto("click", clicarFornecedor)

            forncedorN = "/html/body/div[1]/div/div[3]/form/div/div[4]/table/tbody/tr/td[2]"
            retornaObjeto("text", forncedorN)
            valorItem  = "/html/body/div[1]/div/div[3]/form/div/div[4]/table/tbody/tr/td[6]"
            retornaObjeto("text", valorItem)

            print(f"{nomeOM} + {nLicita} + {qntOM} + {iniVig} + {fimVig}")

            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        else:
            print(f"Quantidade: {qntOM}")
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
    
    comeco += 1        
```
